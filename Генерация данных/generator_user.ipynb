### Считываем данные

```python
user = pd.read_csv('user.csv')
```

```python
names = pd.read_csv('russian_names.csv', sep=';')
```

```python
names = names[['Name', 'Sex']]
```

```python
surnames = pd.read_csv('russian_surnames.csv', sep=';')
```

```python
surnames = surnames[['Surname', 'Sex']]
```

```python
#определяем половую принадлежность фамилии
sex = []
sur = surnames['Surname'].values
i = 0
while i < len(sur)-1:
    if sur[i+1][:-1] == sur[i]:
        sex.append('М')
        sex.append('Ж')
        i += 2
    elif (sur[i+1][:-2] == sur[i][:-2]) and (sur[i][-2:] == 'ая') and (sur[i+1][-2:] == 'ий'):
        sex.append('Ж')
        sex.append('М')
        i += 2
    else:
        sex.append('unisex')
        i+=1
```

```python
sex.append('unisex')
```

```python
surnames.Sex = sex
```

```python
surnames = surnames[surnames.Sex!='unisex']
```
